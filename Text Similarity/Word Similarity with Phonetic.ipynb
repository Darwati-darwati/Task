```python
#import library what we needed
import textdistance
import jellyfish
```

let's make def to make simple code for this case
siphonetic, it means we using phonetic score and levenshtein score
levens, it means we only use levenshtein score

```python

def siphonetic(row):
    kata1 = str(row['kata1'])
    kata2 = str(row['kata2'])
    if jellyfish.metaphone(kata1)==jellyfish.metaphone(kata2):
        scoremetaphone=1
    else:
        scoremetaphone=0
        
    if jellyfish.soundex(kata1)==jellyfish.soundex(kata2):
        scoresoundex=1
    else:
        scoresoundex=0

    phonetic=(scoresoundex+scoremetaphone)/2
    score_levenshtein=textdistance.levenshtein.normalized_similarity(kata1, kata2)
    
    score_similarity=(score_levenshtein+phonetic)/2
    return score_similarity

def levens(row):
    kata1 = str(row['kata1'])
    kata2 = str(row['kata2'])
    return textdistance.levenshtein.normalized_similarity(kata1, kata2)
    
```

### Lets try to excute this logic 
we excute this simple logic to typo dataset from google searching. <br>
there are 39172 typo words and also the right words

```python
import pandas as pd
typo=pd.read_csv('C://Users//Lenovo//Downloads//typo-corpus-r1.txt',sep='\t',header=None)
typo=typo[[0,1]]
typo.columns=['kata1','kata2']
```

```python
%%time
typo['siphonetic']=typo.apply(lambda x: siphonetic(x),axis=1)
```

```python
%%time
typo['levens']=typo.apply(lambda x: levens(x),axis=1)
```

based on excute time, siphonetic need time less than levens. <br>
it's okay, because on siphonetic we using 2 phonetic algorithm, and levens as similarity algoritm
